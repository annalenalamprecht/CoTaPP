<h3 style='color:#3981CB'> Jupyter Notebooks </h3> 

So far we have simply used the IPython console in Spyder or a plain terminal command line as user interfaces for interacting with our Python programs. There are many other variants of user interfaces for Python programs, which for example come in the form of windows-based desktop applications or embedded into web browsers.

Before we look at  “real” graphical user interfaces (GUIs), note that Jupyter Notebooks (https://jupyter-notebook.readthedocs.io/en/stable/) provide a kind of hybrid form of code editor and user interface: Jupyter Notebooks allow for the combination of code and its output along with additional content for documentation in a browser-based interface. They have become extremely popular for data science applications, where users frequently want to change and adapt the data analysis code while interacting with complex data sets.

Anaconda includes Jupyter Notebooks, so we don’t have to install any extra software for using them. Jupyter Notebooks are edited in a browser, but we have to start up the Jupyter Notebook Server first so that the “website” becomes available. To do that, simply type ```jupyter notebook``` into your terminal (as shown on the <mark style="background-color: #FFFF00">first picture </mark>) and press enter. Much of the Jupyter interface is pretty self-explanatory. We will take a quick tour through the interface in the lecture, but you can also easily look up the information in different online sources, e.g. https://jupyter-notebook.readthedocs.io/en/latest/notebook.html#notebook-documents, so we don’t repeat all that in this reader.

The file “Adrien's Mc Donald's exercise (sample solution).ipynb” contains a complete example of a Jupyter Notebook with content you are familiar with

Note that you can use “normal” Python modules within Jupyter, too. That is, for a project that consists of several modules, you don’t have to turn your complete program code into Jupyter notebooks, but only the main .py file. You can simply upload/copy the other .py files to the same directory as the notebook, and import them to the notebook as usual.
